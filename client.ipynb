```python
import jsonimport requestsBASE_URL = "http://localhost:8000"def print_json(response: requests.Response) -> None:    "응답 객체를 보기 좋게 출력합니다."    try:        response.raise_for_status()        payload = response.json()    except Exception as exc:  # noqa: BLE001        print(f"⚠️ 요청 실패: {exc}\n본문: {response.text}")        return    print(f"Status: {response.status_code}")    print(json.dumps(payload, ensure_ascii=False, indent=2))
```
